```python
# Define additional stopwords
ADDITIONAL_STOPWORDS = ['ignore', 'me']

def clean(sentence):
    """
    Cleans a sentence by lemmatizing words, removing stopwords, and normalizing text.
    """
    wnl = nltk.stem.WordNetLemmatizer()
    stopwords = nltk.corpus.stopwords.words('english') + ADDITIONAL_STOPWORDS
    
    # Normalize and lower the sentence
    sentence = (
        unicodedata.normalize('NFKD', sentence)
        .encode('ascii', 'ignore')
        .decode('utf-8', 'ignore')
        .lower()
    )
    
    # Tokenize into sentences, then process each sentence
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', sentence)
    cleaned_sentences = []
    for sent in sentences:
        words = re.sub(r'[^\w\s]', '', sent).split()
        cleaned_words = [wnl.lemmatize(word) for word in words if word not in stopwords]
        cleaned_sentences.append(cleaned_words)
    
    return cleaned_sentences

def add_cleaned_words_column(df, column_name, new_column_name):
    """
    Adds a new column with cleaned words to the dataframe.
    Each row in the new column contains a list of lists (cleaned words for each sentence).
    
            Parameters:
                    df (dataframe): A pandas dataframe
                    column_name (str): The name of the column to clean
                    new_column_name (str): The name of the new column to add

            Returns:
                    df (dataframe): The modified dataframe with the new column
    """
    df[new_column_name] = df[column_name].apply(lambda x: clean(str(x)))
    return df
```

```python
# Add cleaned words column
df = add_cleaned_words_column(df, 'Summary', 'List of Sentences')
```
